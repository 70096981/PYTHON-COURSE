## Ticks & Labels

```python
from mpl.ticker import MultipleLocator as ML
from mpl.ticker import ScalarFormatter as SF
ax.xaxis.set_minor_locator(ML(0, 2))
ax.xaxis.set_minor_formatter(SF())
ax.tick_params(axis='x',which='minor',rotation=90)
```

## Lines & Markers 


```python
x = np.linspace(0.1, 10*np.pi, 1000)
y = np.sin(x)
plt.plot(x, y, 'C1o:', markevery=25, mec='1.0')
```

```python
fig, ax = plt.subplots()
ax.set_xscale('log')
ax.plot(x, y, 'C1o-', markevery=25, mec='1.0')
```

```python
plt.fill_betweenx([-1, 1],[0],[2*np.pi])
plt.text(0, -1, r' period $/phi$')
```

```python
plt.plot(x, np.sin(x), 'C0', label='Sine')
plt.plot(x, np.cos(x), 'C1', label='Cosine')
plt.legend(bbox_to_anchor=(0, 1, 1, .1),ncol=2, mode='expand', loc='lower left')
```

```python
plt.annotate("A", (x[250],y[250]),(x[250],-1),
ha="center", va="center",arrowprops =
{"arrowstyle" : "->", "color": "C1"})
```
